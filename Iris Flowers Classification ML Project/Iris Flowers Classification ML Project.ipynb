```python
corr = df.drop('class', axis=1).corr()
fig, ax = plt.subplots(figsize=(5,4)) #figsize indicates the width and height
#fig is a figure or collection of subplots
# ax is a single subplot
sns.heatmap(corr,annot=True,ax=ax, cmap='coolwarm') # corr is the data
# if values are -1 or 1 it indicates a high correlation
# anything in the range of 0 to 0.1 indicates negligible correlation and 
# also indicates they are independent variables 
```

## LABEL ENCODER

```python
#converts labels to numeric form so as to convert it to machine-readable format
# categorical attributes which is in string format will be asssigned a number
# here the 'class' having Iris-setosa will be assigned a number 0, Iris-versicolor
# a number 1, and Iris-Verginica a number 2
```

```python
!pip install scikit-learn
```

```python
from sklearn.preprocessing import LabelEncoder
#initialisation
le = LabelEncoder()
```

```python
df['class'] = le.fit_transform(df['class']) # converts categorical data to numeric type
df.head()
```

## MODEL TRAINING

```python
from sklearn.model_selection import train_test_split
# train - 70%
# test - 30%
X = df.drop(columns=['class']) # inpit
Y = df['class'] # output
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.30)
```

```python
# Basic classification model 
from sklearn.linear_model import LogisticRegression
model = LogisticRegression() # initialising the model with LR
```

```python
#Training the model with input and output parameters
model.fit(x_train, y_train)
```

```python
#Model testing
#print metrics to get performance
print("Accuracy: ", model.score(x_test, y_test)*100)
# we can get more accuracy by tuning of hyperparameters of logistic regression
# for more accuracy we can shuffle the dataset for testing and training DATASET
```

```python
#Lets import different models 
# knn - k nearest neighbors
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
print("Accuracy: ", model.score(x_test, y_test)*100)
```

```python
# decision tree
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(x_train,y_train)
print("Accuracy: ", model.score(x_test, y_test)*100)
```
